```python
#let's graph this!
import matplotlib.pyplot as plt
plt.style.use('default')
plt.style.use('ggplot')

even=[]
for t in range (3,1000):
    if t%2==0:
        even.append(t)
        
list_len=[]
for n in even:
    list_p=[] #list of primes
    for k in range(2, n):
        if k > 2: 
            for i in range(2, k): 
                if (k % i) == 0: #exclude any composite
                    break
            else: 
                list_p.append(k) #append any primes to list

    list_pc=[] #prime-composite pair
    list_pp=[] #prime-prime pairs
    for p in list_p:
        if n-p != 1:
            for i in range (2, n-p):
                if ((n-p)% i ==0):
                    list_pc.append((p, n-p))
                    break
            else:
                list_pp.append((p, n-p))
    list_len.append(len(list_pp))

plt.scatter(even, list_len, c=even)

plt.xlabel('even #')
plt.ylabel('# of p-p representations')
plt.title('prime prime representations vs even #');
```
